## TF-IDF 

```python
df2.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# 次元数は、max_features=20000
tfidf = TfidfVectorizer(stop_words="english", max_features=10000)
df2["overview"] = df2["overview"].fillna(" ")

tfidf_matrix = tfidf.fit_transform(df2["overview"])
tfidf_matrix.shape
```

```python
tfidf.get_feature_names()[-10:]
```

```python
tfidf.vocabulary_ 
```

```python
# コサイン類似度を求める
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
cosine_sim
```

- 映画のタイトルを入力として受け取り、最も類似している10の映画のリストを出力する関数を定義する.
- そのためには、映画のタイトルとDataFrameインデックスの逆マッピングが必要.

```python
# のちにindexをもとに映画名を呼び出すときに使う
movie2idx = pd.Series(df2.index, index=df2["title"]).drop_duplicates()
```

```python
movie2idx.head(10)
```

```python
# 類似度が高いTop10を返り値にする
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movie2idx[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df2["title"].iloc[movie_indices]
```

```python
# アンハサウェイのマイ・インターン
get_recommendations("The Intern")
```

```python
# プラダを着た悪魔
get_recommendations("The Devil Wears Prada")
```
